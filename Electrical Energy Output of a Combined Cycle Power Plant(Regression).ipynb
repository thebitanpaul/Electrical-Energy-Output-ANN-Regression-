```python
import numpy as np
import pandas as pd
import tensorflow as tf
tf.__version__
```

```python
df = pd.read_excel('/content/Folds5x2_pp.xlsx')
df
```

```python
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.2, random_state=0)
```

##Initialize The ANN

```python
ann = tf.keras.models.Sequential()
```

##Creating Layers Of ANN

Input & 1st Hidden Layer

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

2nd Hidden Layer

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

Output Layer

```python
ann.add(tf.keras.layers.Dense(units=1))
```

we don't need any activation function on the output neuron in case of ANN Regression model

##Training The ANN

Compiling all the layers of ANN

```python
ann.compile(optimizer='adam', loss='mean_squared_error')
```

In case of Continuous output or ANN Regression model, we use loss method = 'mean_squared_error'

Training The ANN

```python
ann.fit(x_train, y_train, batch_size=32, epochs=200)
```

##Predicting

```python
y_pred = ann.predict(x_test)
y_pred
```

```python
np.set_printoptions(precision=2)
```

```python
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test),1)), 1))
```
